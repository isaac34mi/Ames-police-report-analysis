```python
%%html
<script>
    var code_show=true; //true -> hide code at first

    function code_toggle() {
        var btn = document.getElementById("toggleid");
        $('div.prompt').hide(); // always hide prompt

        if (code_show){
            $('div.input').hide();
            btn.text = "Show code";
            
        } else {
            $('div.input').show();
            btn.text = "Hide code";
        }
        code_show = !code_show
    }
    $( document ).ready(code_toggle);
</script>
<style>
.toggle:link, a:visited {
    background-color: #008080;
    color: white;
    padding: 14px 25px;
    text-align: center;
    text-decoration: none;
    border-radius:5px;
    display: inline-block;
}
a:hover, a:active {
    background-color: #008080;
}
</style>
<a class="toggle" id="toggleid" href="javascript:code_toggle()">Hide code</a>
```

```python
from pyecharts import Bar,Pie,Line,Overlap,Page,Scatter
line = Line("Sales")
line.add("Sales by Round $000", roundnum,sales, is_smooth=True)
bar = Bar("", "Data from our Financial Statements")
bar.add("Market Share Low", roundnum, [i*1000 for i in market_share1])
bar.add("Market Share High", roundnum, [i*1000 for i in market_share2])
overlap = Overlap(height=700,width=1200)
overlap.add(line)
overlap.add(bar)
overlap
```
